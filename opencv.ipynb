# image filtering(smooting)

# edge detection

```python
img = cv2.imread(r"D:\downloads\mahesh-babu-on-chiranjeevis-dialogue-in-guntur-karam.jpg")
gaussian_blur = cv2.GaussianBlur(img, (5, 5), 0)
egdes=cv2.Canny(gaussian_blur, 100, 200)
cv2.imshow('edges',egdes)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gaussian_blur = cv2.GaussianBlur(img, (5, 5), 0)
sobelx=cv2.Sobel(gaussian_blur,cv2.CV_64F,1,0,ksize=5)
sobely=cv2.Sobel(gaussian_blur,cv2.CV_64F,0,1,ksize=5)
sobel=cv2.bitwise_or(sobelx,sobely)
cv2.imshow('sobel',sobel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np

blur = cv2.GaussianBlur(img, (5, 5), 0)
# Sobel in X direction
sobelx = cv2.Sobel(blur, cv2.CV_64F, 1, 0, ksize=3)

# Sobel in Y direction
sobely = cv2.Sobel(blur, cv2.CV_64F, 0, 1, ksize=3)

# Convert to absolute values
sobelx = cv2.convertScaleAbs(sobelx)
sobely = cv2.convertScaleAbs(sobely)

# Combine X and Y edges
sobel_combined = cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0)

# -------------------------------
# Display results
# -------------------------------
# cv2.imshow('Original', img)
# cv2.imshow('Canny Edges', canny_edges)
# cv2.imshow('Sobel X', sobelx)
# cv2.imshow('Sobel Y', sobely)
cv2.imshow('Sobel Combined', sobel_combined)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Thresholding

```python
img=cv2.imread(r"D:\downloads\Gemini_Generated_Image_4jrh784jrh784jrh.png")
# simple thresholding
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
_,simple=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
cv2.imshow('simple',simple)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
adaptive=cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
cv2.imshow('adaptive',adaptive)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# otsu's thresholding
_,otsu=cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
cv2.imshow('otsu',otsu)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
